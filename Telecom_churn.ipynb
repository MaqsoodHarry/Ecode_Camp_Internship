### Check Relation between diffrent columns

```python
sns.lmplot(data=new_df, x='MonthlyCharges' ,y='TotalCharges', fit_reg=False)
```

```python
new_df['MonthlyCharges'].corr(new_df['TotalCharges'])
```

this shows positve co_relation between MonthlyCharges and TotalCharges

```python
mth= sns.kdeplot(new_df.MonthlyCharges[(new_df['Churn'] == 1)], color='red', shade=True)
mth= sns.kdeplot(new_df.MonthlyCharges[(new_df['Churn'] == 0)], color='blue', shade=True)
mth.legend(['Churn','No Churn'], loc='upper right')
mth.set_xlabel('Density')
mth.set_ylabel('Monthly Charges')
mth.set_title('Monthly Charges by Churn')
```

This shows high MonthlyCharges More Churn rate
